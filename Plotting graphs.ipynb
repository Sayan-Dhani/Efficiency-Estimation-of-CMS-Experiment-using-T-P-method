```python
# Draw the histogram
root.gStyle.SetOptStat(1)
root.gStyle.SetTextFont(42)

c1 = root.TCanvas('c1', 'c1', 1000, 600)
c1.SetLogy()
c1.SetGrid()

Z_mass_21.SetTitle("Invariant Mass of Tag & Probe Pair")
Z_mass_21.GetXaxis().SetTitleSize(0.04)
Z_mass_21.GetYaxis().SetTitleSize(0.04)

Z_mass_21.GetXaxis().SetTitle("Mass(M_mu_mu)(GeV)")
Z_mass_21.GetYaxis().SetTitle("No. of Events")

Z_mass_21.Draw()
c1.Draw()

# Print histogram statistics
hist_name = Z_mass_21.GetName()
entries = Z_mass_21.GetEntries()
mean = Z_mass_21.GetMean()
std_dev = Z_mass_21.GetStdDev()
underflow = Z_mass_21.GetBinContent(0)
overflow = Z_mass_21.GetBinContent(Z_mass_21.GetNbinsX() + 1)
integral = Z_mass_21.Integral()
skewness = Z_mass_21.GetSkewness()
kurtosis = Z_mass_21.GetKurtosis()

print("Histogram Name:", hist_name)
print("Entries:", entries)
print("Mean:", mean)
print("Std Dev:", std_dev)
print("Underflow:", underflow)
print("Overflow:", overflow)
print("Integral:", integral)
print("Skewness:", skewness)
print("Kurtosis:", kurtosis)

```
