```python
start = time.time()    # note the starting time

for i,ent in enumerate(chain):
    
    ### ================================= Event number block ========================================== ###
    #if i>1000000: break;
    if (i > 0 and i%1000000 == 0): print("Processing: ", i, "th event!")
    
    ### ================================================================================================ ###
    ### ==================================== MAJOR SELECTION BLOCK ===================================== ###
    ### ================================================================================================ ###
    
    ### ================================= Primary selection of events ================================== ###
    if ent.HLT_IsoMu24_eta2p1 !=1: continue;
    if ent.MET_pt >40 : continue;
    if ent.nMuon <2 : continue;
    
    # Muon properties block
    u_pt       = list(ent.Muon_corrected_pt)
    u_eta      = list(ent.Muon_eta)
    u_phi      = list(ent.Muon_phi)
    u_mass     = list(ent.Muon_mass)
    u_charge   = list(ent.Muon_charge)
    u_dxy      = list(ent.Muon_dxy)
    u_dz       = list(ent.Muon_dz)
    u_iso      = list(ent.Muon_miniPFRelIso_all)
    u_tightId  = list(ent.Muon_tightId)
    u_mediumId = list(ent.Muon_mediumId)
    u_looseId  = list(ent.Muon_looseId)
    
    
    ### ===================================  Tag Selection Block ======================================= ###
    tagid = []
    for k in range(ent.nMuon):
        if(u_pt[k]> 24 and abs(u_eta[k]) < eta_max and u_iso[k] < 0.3 and u_tightId[k]==1): tagid.append(k)

    # Require at least 1 tag muon
    if len(tagid) < 1: continue;
        
    # randomize the tagid list
    if len(tagid) > 1: random.shuffle(tagid)
    tidx = tagid[0]
        
    # get four vector of the tag muon
    u1LV = root.TLorentzVector()
    u1LV.SetPtEtaPhiM(u_pt[tidx],u_eta[tidx],u_phi[tidx],u_mass[tidx])
    
 
    ### ===================================  Probe Selection Block ===================================== ###
    probeid = []
    for k in range(ent.nMuon):
        if k == tidx: continue
        if (u_charge[tidx]*u_charge[k] > 0): continue;
        if abs(u_eta[k]) > eta_max: continue
            
        u2LV = root.TLorentzVector()
        u2LV.SetPtEtaPhiM(u_pt[k],u_eta[k],u_phi[k],u_mass[k])

        Z_LV = u1LV + u2LV
        mass = Z_LV.M()
        Z_mass_21.Fill(mass)     # fill the invariant mass the Tag & Probe Pair
        if 20 < mass < 160 : probeid.append(k)

    if len(probeid) < 1: continue

    # do not randomize probeid (for now)
    # if len(probeid) > 1: random.shuffle(probeid)
    pidx = probeid[0]
    u2LV = root.TLorentzVector()
    u2LV.SetPtEtaPhiM(u_pt[pidx],u_eta[pidx],u_phi[pidx],u_mass[pidx])

    deltaRTPH.Fill(u1LV.DeltaR(u2LV))   # fill the delta angle b/w the Tag & Probe
    
    
    ### =================================== Electron Veto Block ========================================== ###
    ele_pt    = list(ent.Electron_pt)
    ele_eta   = list(ent.Electron_eta)
    ele_phi   = list(ent.Electron_phi)
    ele_mass  = list(ent.Electron_mass)
    ele_mvaid = list(ent.Electron_mvaFall17V2Iso_WPL)
    
    eleid = []
    for j in range(ent.nElectron):
        if ele_pt[j]<10: continue
        if abs(ele_eta[j]) > 2.5: continue
        if ele_mvaid[j]==0: continue
        ele_LV = root.TLorentzVector()
        ele_LV.SetPtEtaPhiM(ele_pt[j], ele_eta[j], ele_phi[j], ele_mass[j])            
        if (u1LV.DeltaR(ele_LV)<0.4 or u2LV.DeltaR(ele_LV)<0.4): continue
        eleid.append(j)        
    # veto the event if there is a good electron
    nEleH.Fill(len(eleid))
    if len(eleid) > 0: continue
    
    
    ### ================================= Jet & bJet Veto Block ========================================== ###
    jet_pt   = list(ent.Jet_pt)
    jet_eta  = list(ent.Jet_eta)
    jet_phi  = list(ent.Jet_phi)
    jet_mass = list(ent.Jet_mass)
    jet_btag = list(ent.Jet_btagDeepFlavB)
    jetid  = []
    bjetid = []
    for j in range(ent.nJet):
        if jet_pt[j] < 30: continue
        if abs(jet_eta[j])> 4.7: continue # this is based on the calorimeter
        jet_LV = root.TLorentzVector()
        jet_LV.SetPtEtaPhiM(jet_pt[j], jet_eta[j], jet_phi[j], jet_mass[j])
        if (u1LV.DeltaR(jet_LV)<0.4 or u2LV.DeltaR(jet_LV)<0.4): continue
        
        jetid.append(j)

        if abs(jet_eta[j]) > 2.5: continue 
        if jet_btag[j] <= 0.33: continue
        bjetid.append(j)

    nJetH.Fill(len(jetid))
    nbJetH.Fill(len(bjetid))
    if len(jetid) > 0: hJetPtH.Fill(jet_pt[jetid[0]])

    # veto the event if there are good Jet
    if len(jetid) > 3: continue
    #if len(jetid) > 0 and jet_pt[jetid[0]] > 60: continue;

    # veto the event if there is a bJet
    if len(bjetid) > 0: continue
    
    ### ================================================================================================== ###
    ### ===================================== MAJOR PLOTTING BLOCK ======================================= ###
    ### ================================================================================================== ###
    
    ### =================================== NO CUT ON PROBE :: CUT 0 ===================================== ### 
    
    mu_pt_c0_21.Fill(u_pt[pidx])   # fill probe pT (cut 0)
    
    
    probe_pt_iso_c0_21.Fill(u_pt[pidx],u_iso[pidx])  # 2D Histogram : pT v/s isolation
    probe_iso_pt_c0_21.Fill(u_iso[pidx],u_pt[pidx])  # 2D Histogram : isolation v/s pT


    probe_pt_iso_c0PH_21.Fill(u_pt[pidx],u_iso[pidx],1)  # 2D Profile Histogram : pT v/s isolation
    probe_iso_pt_c0PH_21.Fill(u_iso[pidx],u_pt[pidx],1)  # 2D Profile Histogram : isolation v/s pT
    
    # Filling Probe pT Histograms for different ranges of isolation 
    if u_iso[pidx] < 0.10:
        mu_pt_iso010_c0_21.Fill(u_pt[pidx])
    if u_iso[pidx] < 0.15:
        mu_pt_iso015_c0_21.Fill(u_pt[pidx])
    if u_iso[pidx] < 0.20:
        mu_pt_iso020_c0_21.Fill(u_pt[pidx])
    if u_iso[pidx] < 0.25:
        mu_pt_iso025_c0_21.Fill(u_pt[pidx])
    if u_iso[pidx] < 0.30:
        mu_pt_iso030_c0_21.Fill(u_pt[pidx])
    if u_iso[pidx] < 0.35:
        mu_pt_iso035_c0_21.Fill(u_pt[pidx])
    if u_iso[pidx] < 0.40:
        mu_pt_iso040_c0_21.Fill(u_pt[pidx])
        
    # Filling Tag & Probe Pair inv. mass Histograms for different ranges of Probe pT
    if 0 < u_pt[pidx] < 20:
        Zmc0ptr_0_20.Fill(mass)
    elif 20 < u_pt[pidx] < 30:
        Zmc0ptr_20_30.Fill(mass)
    elif 30 < u_pt[pidx] < 40:
        Zmc0ptr_30_40.Fill(mass)
    elif 40 < u_pt[pidx] < 50:
        Zmc0ptr_40_50.Fill(mass)
    elif 50 < u_pt[pidx] < 60:
        Zmc0ptr_50_60.Fill(mass)
    elif 60 < u_pt[pidx] < 80:
        Zmc0ptr_60_80.Fill(mass)
    elif 80 < u_pt[pidx] < 200:
        Zmc0ptr_80_200.Fill(mass)

    ### =========================== RECONSTRUCTION CUT ON PROBE :: CUT 1 ================================= ### 
    
    # apply the tight id cut
    if u_tightId[pidx] == 1:
        mu_pt_tc1_21.Fill(u_pt[pidx])   # fill probe pT (cut 1)
        if u_iso[pidx] < 0.3:
            mu_pt_tc2_21.Fill(u_pt[pidx]) # fill probe pT (cut 1)
    
    # apply the loose id cut        
    if u_looseId[pidx] == 1:
        mu_pt_lc1_21.Fill(u_pt[pidx])   # fill probe pT (cut 1)
        if u_iso[pidx] < 0.3:
            mu_pt_lc2_21.Fill(u_pt[pidx]) # fill probe pT (cut 1)  
            
            
    # apply the medium id cut       
    if u_mediumId[pidx] == 1:
        mu_pt_mc1_21.Fill(u_pt[pidx])   # fill probe pT (cut 1)
        
        probe_pt_iso_c1_21.Fill(u_pt[pidx],u_iso[pidx])   # 2D Histograms : pT v/s isolation
        probe_iso_pt_c1_21.Fill(u_iso[pidx],u_pt[pidx])   # 2D Histograms : isolation v/s pT
        
        probe_pt_iso_c1PH_21.Fill(u_pt[pidx],u_iso[pidx],1)   # 2D Profile Histograms : pT v/s isolation
        probe_iso_pt_c1PH_21.Fill(u_iso[pidx],u_pt[pidx],1)   # 2D Profile Histograms : isolation v/s pT
        
        # Filling Probe pT Histograms for different ranges of isolation 
        if u_iso[pidx] < 0.10:
            mu_pt_iso010_c1_21.Fill(u_pt[pidx])
        if u_iso[pidx] < 0.15:
            mu_pt_iso015_c1_21.Fill(u_pt[pidx])
        if u_iso[pidx] < 0.20:
            mu_pt_iso020_c1_21.Fill(u_pt[pidx])
        if u_iso[pidx] < 0.25:
            mu_pt_iso025_c1_21.Fill(u_pt[pidx])
        if u_iso[pidx] < 0.30:
            mu_pt_iso030_c1_21.Fill(u_pt[pidx])
        if u_iso[pidx] < 0.35:
            mu_pt_iso035_c1_21.Fill(u_pt[pidx])
        if u_iso[pidx] < 0.40:
            mu_pt_iso040_c1_21.Fill(u_pt[pidx])
        
        # Filling Tag & Probe Pair inv. mass Histograms for different ranges of Probe pT
        if 0 < u_pt[pidx] < 20:
            Zmc1ptr_0_20.Fill(mass)
        elif 20 < u_pt[pidx] < 30:
            Zmc1ptr_20_30.Fill(mass)
        elif 30 < u_pt[pidx] < 40:
            Zmc1ptr_30_40.Fill(mass)
        elif 40 < u_pt[pidx] < 50:
            Zmc1ptr_40_50.Fill(mass)
        elif 50 < u_pt[pidx] < 60:
            Zmc1ptr_50_60.Fill(mass)
        elif 60 < u_pt[pidx] < 80:
            Zmc1ptr_60_80.Fill(mass)
        elif 80 < u_pt[pidx] < 200:
            Zmc1ptr_80_200.Fill(mass)

    ### ================= RECONSTRUCTION & ISOLATION CUT ON PROBE :: CUT 2 (with Midium ID)================ ### 

        if u_iso[pidx] < 0.45:
            mu_pt_mc2_21.Fill(u_pt[pidx]) # fill probe pT (cut 1)
            
            # Filling Tag & Probe Pair inv. mass Histograms for different ranges of Probe pT
            if 0 < u_pt[pidx] < 20:
                Zmc2ptr_0_20.Fill(mass)
            elif 20 < u_pt[pidx] < 30:
                Zmc2ptr_20_30.Fill(mass)
            elif 30 < u_pt[pidx] < 40:
                Zmc2ptr_30_40.Fill(mass)
            elif 40 < u_pt[pidx] < 50:
                Zmc2ptr_40_50.Fill(mass)
            elif 50 < u_pt[pidx] < 60:
                Zmc2ptr_50_60.Fill(mass)
            elif 60 < u_pt[pidx] < 80:
                Zmc2ptr_60_80.Fill(mass)
            elif 80 < u_pt[pidx] < 200:
                Zmc2ptr_80_200.Fill(mass)
            
    
print ("Loop is completed")     # LOOP COMPLETTION MASSAGE
end = time.time()   # Note the end time
print(end-start)    # Show the time taken in seconds to rum the code
RootfileWrite()
```
